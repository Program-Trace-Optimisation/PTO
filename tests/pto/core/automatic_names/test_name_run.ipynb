# Tests for automatic names run

```python
from pto.core.automatic_names import _run as run, rnd
```

```python
def random_program():
    return([rnd.random(),
            rnd.choice(['a','b','c']),
            rnd.randint(1, 10)])

def fitness(sol): return sol[0]
```

```python
run(random_program, fitness, name_type='str') # expect error as program not annotated
```

```python
run(random_program, fitness, name_type='lin')
```
