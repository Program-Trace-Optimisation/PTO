# Tests for hill_climber

```python
from pto.solvers import HC

from pto.core.base import Op, tracer, Dist

import random
```

```python

def random_program():
    return([tracer.sample('pos 1', Dist(random.random)),
            tracer.sample('pos 2', Dist(random.choice, ['a','b','c'])),
            tracer.sample('pos 3', Dist(random.randint, 1, 10))])

def fitness(sol): return sol[0]
```

```python
op = Op(generator=random_program, fitness=fitness)
```

```python
# instantiate hill climber
hc=HC(op)
```

```python
# hill climber current parameters
hc.__dict__
```

```python
# execute search
res=hc()
```

```python
# result
print('sol:    ', res[0].pheno)
print('trace:  ', res[0].geno)
print('fitness:', res[1])
```

```python
# test callback
from pprint import pprint

hc=HC(op, callback=print)

pprint(hc.__dict__)

res=hc()
```

```python
# test callback break

count=0
def maxit(_):
    global count
    count += 1
    print(count)
    if count >= 10:
        print('stop!')
        return True
    
hc=HC(op, callback=maxit)

pprint(hc.__dict__)

res=hc()

print('sol:    ', res[0].pheno)
print('trace:  ', res[0].geno)
print('fitness:', res[1])
```
